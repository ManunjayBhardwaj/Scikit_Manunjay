```python
#Linear Regression Model

# Predict a student's final exam score based on the number of hours they study.
```

```python
import numpy as np # as the exam score and hour study has numeric integer values we import numpy
import pandas as pd # as the two values create a dataframe
from sklearn.model_selection import train_test_split # we don't directly import the whole sklearn library because it loads all the modules with it making the program very heavy
from sklearn.linear_model import LinearRegression # this helps us import linear regression model from our sklearn library
```

```python
# step 2
data={'Hours_Study':[2,3,4,5,6,7,8,9,10],'Exam_Score':[50,60,70,75,80,85,90,95,100]} # as we can see that the exam score linearly increases with the hours the person has studied therefore we have used LinearRegression
```

```python
# step 3
# creating a data frame from the above data
df=pd.DataFrame(data)
```

```python
# step 4
# now we have to extract the features from the data so that we can train test
# This is called Feature Extraction
X=df[["Hours_Study"]]
Y=df[["Exam_Score"]]
# we can see that we have extracted the features to make the train test split
# we have used a 2-D array to extract because it is a dataframe
```

```python
# step 5
# in train and test method : we train 80 percent of the data and match 20 percent of the data
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
# random state is a parameter set to 42 to assure the reproducibility of the event.Meaning whenever we put the random state to 42 we ensure the reproducibility of the data meaning whenever we train the data it will give the same value leading to better training
```

```python
# step 6
model=LinearRegression()# we can directly use the linearregression too but we store it in a variable called model so that we don't have to write it again and again
# we create an instance of the LinearRegression model so that we can call it whenever we need it in the form of the variable: model
```

```python
# step 7
# now we have to fit the model for training and testing
model.fit(x_train,y_train)

```

```python
# step 8
# User input testing so that we can test any data
user_input=float(input("Enter the number of hours you study: "))

# step 9
predicted_score=model.predict([[user_input]])
# we had the data 2-D so we have to input 2-D

# printing the output
print("Predicted exam score is : ",predicted_score)
```

```python
# now we plot the data to check relationship:-
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x_test,y_test)
fig.suptitle('Hours Study vs Exam Score')
ax.set_xlabel('Hours Study')
ax.set_ylabel('Exam Score')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
# Now we will try multiple linear regression problem
# Problem statement: A company wants to predict the monthly sales (in thousands of units) of a product based on three factors:Advertising spend (in thousands of dollars).Number of sales calls made by the sales team.Discount percentage offered on the product.
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Dataset as a dictionary
data = {
    "Advertising Spend ($k)": [10, 15, 20, 25, 30, 35, 40, 45],
    "Sales Calls": [50, 60, 55, 65, 70, 80, 85, 90],
    "Discount (%)": [5, 10, 15, 5, 10, 15, 10, 5],
    "Monthly Sales (k units)": [20, 22, 25, 28, 30, 35, 37, 40]
}
# creating a data frame
df=pd.DataFrame(data)

# now feature extraction ( independent variable)
X=df[["Advertising Spend ($k)","Sales Calls", "Discount (%)"]]

# now we extract the target variable
Y=df[["Monthly Sales (k units)"]]

# train and test split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# importing the model
model = LinearRegression()
model.fit(x_train,y_train)

# taking the user input
advertising_spend = float(input("Enter the number of Advertising Spend (k): "))
sales_calls = float(input("Enter the number of Sales Calls (k): "))
discount = float(input("Enter the discount (%): "))

input_data=[[advertising_spend, sales_calls, discount]]
prediction = int(model.predict(input_data))
print(f"The monthly sales predicted is:{prediction} k units")
```
